## Step 8

```python
df = fert.merge(pop.merge(life2cont))
df = df.dropna()
```

```python
df
```

Create plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
year_span = []
for n in range(1800,1950,10):
    year_span.append(n)
for n in range(1950,2016):
    year_span.append(n)
year_span
```

```python
for i in year_span:
    boolean_mask = df['year']==i  
    data = df[boolean_mask]
    data.to_csv('df/data' + str(i) + '.csv')
    col = []
    for entry in df[boolean_mask]['continent'].index:
        if df[boolean_mask]['continent'][entry] == 'Asia':
            col.append('red')
        elif df[boolean_mask]['continent'][entry] == 'Europe':
            col.append('blue')
        elif df[boolean_mask]['continent'][entry] == 'Africa':
            col.append('yellow')
        elif df[boolean_mask]['continent'][entry] == 'North America':
            col.append('orange')
        elif df[boolean_mask]['continent'][entry] == 'South America':
            col.append('green')  
        elif df[boolean_mask]['continent'][entry] == 'Australia and Oceania':
            col.append('black')
    plt.figure(figsize=(14,10))
    plt.axis(xmin = 20, xmax = 90, ymin = 0, ymax = 9)
    plt.scatter(x=df[boolean_mask]['life_expectancy'], y=df[boolean_mask]['fertility_rate'], s = df[boolean_mask]['population']/100000, c = col)
    plt.title('Asia: red, Europe: blue, Africa: yellow, North America: orange, South America: green, Australia and Oceania: black')
    plt.xlabel('Life expectancy')
    plt.ylabel('Fertility')
    plt.suptitle('Fertility/Life expectancy of the continents in ' + str(i))
    plt.savefig('plots/solid/' + str(i) + 'solid.png')
    plt.close()
```
